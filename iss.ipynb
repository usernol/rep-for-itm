```python
import requests
import time

all_data = []
latitudes = []
longitudes = []
timestamps = []

for i in range (102): #ping for 17 mins w 10 second interval
    print(i)
    res = requests.get('http://api.open-notify.org/iss-now.json')
    data = res.json()
    latitude = (data['iss_position']['latitude'])
    longitude = (data['iss_position']['longitude'])
    timestamp = (data['timestamp'])
    latitudes.append(latitude)
    longitudes.append(longitude)
    timestamps.append(timestamp)
    all_data.append((latitude,longitude,timestamp))
    time.sleep(10) # 10 seconds interval
```

```python
import matplotlib
from matplotlib import pyplot as plt

plt.scatter(latitudes, longitudes, c=timestamps, cmap = "coolwarm", s = 10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('International Space Station Path')
plt.show()
```
